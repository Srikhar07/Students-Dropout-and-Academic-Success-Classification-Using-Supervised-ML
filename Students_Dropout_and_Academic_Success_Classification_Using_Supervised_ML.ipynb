###Features and target

```python
x = dataset.drop('Target', axis=1)
y = dataset['Target']
```

```python
x.head()
```

```python
y.head()
```

###Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

###Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

###Balancing the dataset

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
x_train_resampled, y_train_resampled = ros.fit_resample(x_train, y_train)
```

```python
counts = y_train_resampled.value_counts()
print(counts)
```

```python
sns.countplot(y = y_train_resampled)
plt.show()
```

###Model training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

models = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Support Vector Machine': SVC(kernel='rbf', random_state=42),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors= 10, metric = 'minkowski'),
    'Random Forest': RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=42),
    'Decision Tree': DecisionTreeClassifier(criterion='entropy', random_state=42)
}
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

for model_name, model in models.items():
  print(f"Training: {model_name}")
  model.fit(x_train_resampled, y_train_resampled)
  print(f"{model_name} trained!")

  print(f"Predicting for training set: {model_name}")
  y_pred_train = model.predict(x_train_resampled)
  print(f"{model_name} training predicted!")

  print(f"Predicting for test set: {model_name}")
  y_pred_test = model.predict(x_test)
  print(f"{model_name} test predicted!")

  #training set results
  print(f"\nTraining set results: {model_name}")
  model_train_accuracy = accuracy_score(y_train_resampled, y_pred_train)
  model_train_f1 = f1_score(y_train_resampled, y_pred_train, average='weighted')
  model_train_precision = precision_score(y_train_resampled, y_pred_train, average='weighted')
  model_train_recall = recall_score(y_train_resampled, y_pred_train, average='weighted')

  print(f"Accuracy: {model_train_accuracy}")
  print(f"F1 Score: {model_train_f1}")
  print(f"Precision: {model_train_precision}")
  print(f"Recall: {model_train_recall}")
  print("-----------------------------------------------------------")

  #test set results
  print(f"\nTest set results: {model_name}")
  model_test_accuracy = accuracy_score(y_test, y_pred_test)
  model_test_f1 = f1_score(y_test, y_pred_test, average='weighted')
  model_test_precision = precision_score(y_test, y_pred_test, average='weighted')
  model_test_recall = recall_score(y_test, y_pred_test, average='weighted')

  print(f"Accuracy: {model_test_accuracy}")
  print(f"F1 Score: {model_test_f1}")
  print(f"Precision: {model_test_precision}")
  print(f"Recall: {model_test_recall}")
  print("===========================================================")
```

###Check overfit

```python
for model_name, model in models.items():
  print(f"model name: {model_name}")
  training_score = model.score(x_train_resampled, y_train_resampled)
  test_score = model.score(x_test, y_test)

  print(f"{model_name} Training Score: {training_score}")
  print(f"{model_name} Test Score: {test_score}")

  gap = training_score - test_score
  print(f"The gap is: {gap}")
  if gap <=0.05:
    print("The Model is Good no overfit!")
  elif gap <=0.1:
    print("The Model is Overfit!")
  else:
    print("The Model is Underfit!")

  print("-----------------------------------------------------------")
```

###Final model

```python
log_model = models['Logistic Regression']
logestic_pred = log_model.predict(x_test)
```

###Confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, logestic_pred)
sns.heatmap(cm, annot=True)
plt.show()
```

###Classification report

```python
from sklearn.metrics import classification_report
classification_report = classification_report(y_test, logestic_pred)
print(classification_report)
```

###Cross validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = log_model, X = x_train, y = y_train, cv = 10)
print(f"The average accuracies among 10 epochs is: {accuracies.mean()}")
print(f"The standard deviation is: {accuracies.std()}")
```
